# Visualization Notebook

This notebook contains visual analysis for lead time, segments, suppliers, risk levels, and recommended actions.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clean_path = "/home/pratyush_device/Documents/college/AIML lab/Project/customer_cleaned.csv"
risk_path = "/home/pratyush_device/Documents/college/AIML lab/Project/risk_scored_orders.csv"
action_path = "/home/pratyush_device/Documents/college/AIML lab/Project/action_recommendations.csv"

df = pd.read_csv(clean_path)
risk_df = pd.read_csv(risk_path)
action_df = pd.read_csv(action_path)

print("Cleaned data shape:", df.shape)
print("Risk data shape:", risk_df.shape)
print("Action data shape:", action_df.shape)
```

```python
# 1) Lead time distribution
plt.figure(figsize=(8, 5))
plt.hist(df["lead_time_days"], bins=20, edgecolor="black")
plt.title("Lead Time Distribution")
plt.xlabel("Lead Time (days)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 2) Market segment average lead time
segment_cols = [col for col in df.columns if col.startswith("market_segment_")]

if segment_cols:
    segment_rows = []
    for col in segment_cols:
        label = col.replace("market_segment_", "")
        avg_lead = df.loc[df[col] == 1, "lead_time_days"].mean()
        segment_rows.append({"Segment": label, "AvgLeadTime": avg_lead})

    segment_summary = pd.DataFrame(segment_rows).sort_values("AvgLeadTime", ascending=False)

    plt.figure(figsize=(9, 5))
    plt.bar(segment_summary["Segment"], segment_summary["AvgLeadTime"])
    plt.title("Average Lead Time by Market Segment")
    plt.xlabel("Market Segment")
    plt.ylabel("Average Lead Time (days)")
    plt.xticks(rotation=20)
    plt.tight_layout()
    plt.show()

    segment_summary
else:
    print("No market segment columns found.")
```

```python
# 3) Supplier impact on order value (top suppliers)
supplier_cols = [col for col in df.columns if col.startswith("supplier_id_")]

if supplier_cols:
    supplier_rows = []
    for col in supplier_cols:
        label = col.replace("supplier_id_", "")
        avg_value = df.loc[df[col] == 1, "order_value_usd"].mean()
        supplier_rows.append({"Supplier": label, "AvgOrderValue": avg_value})

    supplier_summary = pd.DataFrame(supplier_rows).sort_values("AvgOrderValue", ascending=False)

    plt.figure(figsize=(9, 5))
    plt.bar(supplier_summary["Supplier"], supplier_summary["AvgOrderValue"])
    plt.title("Average Order Value by Supplier")
    plt.xlabel("Supplier")
    plt.ylabel("Average Order Value (USD)")
    plt.xticks(rotation=20)
    plt.tight_layout()
    plt.show()

    supplier_summary
else:
    print("No supplier columns found.")
```

```python
# 4) Risk level distribution
risk_counts = risk_df["Risk_Level"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(risk_counts.values, labels=risk_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Risk Level Distribution")
plt.tight_layout()
plt.show()

risk_counts
```

```python
# 5) Action priority workload
priority_counts = action_df["Priority"].value_counts().sort_index()

plt.figure(figsize=(7, 5))
plt.bar(priority_counts.index, priority_counts.values)
plt.title("Action Priority Distribution")
plt.xlabel("Priority")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()

priority_counts
```

```python
# 6) Top high-risk orders by delay probability
top_risk = action_df.sort_values("Delay_Probability", ascending=False).head(15)

plt.figure(figsize=(10, 5))
plt.bar(range(len(top_risk)), top_risk["Delay_Probability"])
plt.title("Top 15 Orders by Delay Probability")
plt.xlabel("Order Rank")
plt.ylabel("Delay Probability")
plt.tight_layout()
plt.show()

top_risk[["Delay_Probability", "Risk_Level", "Priority", "Owner", "Action"]]
```
